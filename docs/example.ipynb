# Example usage

To use `diversedata` in a project, import the package:

```python
import diversedata as dd
```

To list the available data sets:

```python
dd.list_available_datasets()
```

To view the documentation of a specific dataset:

```python
dd.print_data_description('wildfire')
```

To load a specific dataset:

```python
df = dd.load_data("wildfire")
df.head()
```

Explore more in depth examples [here](https://diverse-data-hub.github.io/).
